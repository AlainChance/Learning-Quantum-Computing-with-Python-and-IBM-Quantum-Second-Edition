```python
#Create a simple circuit with just one qubit
qc = QuantumCircuit(1)
```

```python
from qiskit.quantum_info import Statevector
stateVectorResult = Statevector(qc)

print('state vector results ', stateVectorResult)
```

```python
# Added by Alain Chancé
display(array_to_latex(stateVectorResult, prefix="\\text{state vector results} = ", precision = 1))
stateVectorResult.draw('latex')
```

```python
#Display the Bloch sphere
stateVectorResult.draw('bloch')
```

Figure 5.3: Qubit Bloch sphere state vector initialized to $|0\rangle$

```python
#Display the qsphere
stateVectorResult.draw('qsphere')
```

Figure 5.4: Qubit state vector initialized to with phase = 0

```python
qc = QuantumCircuit(1)
qc.x(0)
#Execute the circuit and display results
stateVectorResult = Statevector(qc)
```

```python
# Added by Alain Chancé
display(array_to_latex(stateVectorResult, prefix="\\text{state vector results} = ", precision = 1))
stateVectorResult.draw('latex')
```

```python
#Display the Bloch sphere
stateVectorResult.draw('qsphere')
```

Figure 5.5: Qubit state vector set to $|1\rangle$ with phase = 0

```python
qc = QuantumCircuit(1)
qc.h(0)
#Execute the circuit and display results
stateVectorResult = Statevector(qc)
```

```python
# Added by Alain Chancé
display(array_to_latex(stateVectorResult, prefix="\\text{state vector results} = ", precision = 4))
stateVectorResult.draw('latex')
```

```python
#Display the Bloch sphere
stateVectorResult.draw('bloch')
```

Figure 5.6: Bloch sphere superposition representation, a linear combination of $|0\rangle$ and $|1\rangle$

```python
stateVectorResult.draw('qsphere')
```
